```python

booklist = [[],[],[]]
while True:
    choice = int(input("""
      Welcome to the CRUD management system for the library of Alexandria!.
      What would you like to do today? 
      
      1. Add book
      2. Edit book
      3. view book
      4. Delete book
      
      Enter your choice: 
      """))

    if choice == 1:
        while True:
            class book():
                def __init__(self, title, author, pages):
                    self.author = author
                    self.pages = pages
                    self.title = title
                def generalDescription():
                    print(f"This is an automated message from the library of Alexandria!, Have a nice day")

            bookAdd = int(input("""what is the book type? 
                1. Graphic Novel
                2. Comic book
                3. Magazine
                4. Continue
                """))
            if bookAdd == 1:
                name = input("what is the name of the author? ")
                pages = int(input("how many pages constitute this book? "))
                title = input("what is the title of this book? ")
                coloredQ = input(" is this graphic novel colored? ")
                class GraphicNovel(book):
                    def __init__(self, author = "NA", title = "NA", pages = "NA", coloredQ = "NA"):
                        super().__init__(author, pages, title)
                        self.coloredQ = coloredQ
                    def description(self):
                        super().generalDescription()
                        print(f"""The graphic novel, {self.title} written by {self.author}, it has {self.pages} pages. More details below
                                    colored? = {self.coloredQ}
                            """)
                booklist[0].append(GraphicNovel(author=name, pages=pages, coloredQ=coloredQ,title=title))
                    
            elif bookAdd == 2:
                name = input("what is the name of the author? ")
                pages = int(input("how many pages constitute this book? "))
                title = input("what is the title of this book? ")
                coloredQ = input(" is this comic book colored? ")
                publisher = input("Who is the publisher of the comic book? ")
                class comicBook(book):
                    def __init__(self, author = "NA", title = "NA", pages = "NA", coloredQ = "NA", publisher = "NA"):
                        super().__init__(author, pages, title)
                        self.coloredQ = coloredQ
                        self.publisher = publisher
                        self.title = title
                    def description(self):
                        super().generalDescription()
                        print(f"""This comic book, {self.title} written by {self.author}, it has {self.pages} pages. More details below
                                    colored? = {self.coloredQ}
                                    publisher? = {self.publisher}
                        """)

                booklist[1].append(comicBook(author=name, pages=pages, coloredQ=coloredQ, publisher=publisher,title=title))
            elif bookAdd == 3:
                author = input("what is the author of the author? ")
                pages = int(input("how many pages constitute this book? "))
                title = input("what is the title of this book? ")
                publisher = input("Who is the publisher of this Magazine? ")
                About = (input("What is this magazine about? "))
                class magazine(book):
                    def __init__(self, author = "NA", pages =  "NA", publisher = "NA", About = "NA", title = "NA"):
                        super().__init__(author, pages, title)
                        self.publisher = publisher
                        self.title = title
                        self.About = About
                    def description(self):
                        super().generalDescription()
                        print(f"""This magazine, {self.title} written by {self.author}, it has {self.pages} pages. More details below
                                    About? = {self.About}
                                    publisher? = {self.publisher}""")
                booklist[2].append(magazine(author=author, pages=pages, publisher=publisher, About=About, title = title))
            elif bookAdd == 4:
                backTrace = input("would you like to go back to menu? (Y/N) ")
                if backTrace == "N":
                    continue
                elif backTrace == "Y":
                    break
                else:
                    print("Err")    
            
            else:
                print("not an option, please retry")
                continue
    elif choice == 2:
        print(f"Graphic novels:")
        for i in range(len(booklist[0])-1) :
            print(f"{i+1}., {booklist[0][i].title}, {booklist[0][i].name}")
        for l in range(len(booklist[1])-1):
            print("Comic Books:")
            print(f"{l+1}., {booklist[1][l].title}, {booklist[1][l].name}")
        for n in range(len(booklist[2])-1):
            print("Magazine Books:")
            print(f"{n+1}., {booklist[2][n].title}, {booklist[2][n].name}")
              
```
